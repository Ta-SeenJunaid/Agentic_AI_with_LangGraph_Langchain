## FAISS

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter          

loader = TextLoader("../Data/random/text.txt")
text_doc = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=600, chunk_overlap=100)
docs = text_splitter.split_documents(text_doc)
```

```python
docs
```

```python
embeddings = OllamaEmbeddings(model="llama3:latest", base_url="http://localhost:11434")
vector_store = FAISS.from_documents(docs, embeddings)
vector_store
```

```python
query = "Where a delivery truck was parked?"
docs=vector_store.similarity_search(query)
docs
```

```python
docs[0].page_content
```

```python
retriever = vector_store.as_retriever()
docs=retriever.invoke(query)
docs[0].page_content
```

```python
docs_and_score=vector_store.similarity_search_with_score(query)
docs_and_score
```

```python
embedding_vector = embeddings.embed_query(query)
embedding_vector
```

```python
docs_score=vector_store.similarity_search_with_score_by_vector(embedding_vector)
docs_score
```

```python
vector_store.save_local("../Data/ignore/faiss_index")
```

```python
new_vector_store = FAISS.load_local("../Data/ignore/faiss_index", embeddings, allow_dangerous_deserialization=True)
docs=new_vector_store.similarity_search(query)
docs
```

## ChromaDB

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("../Data/random/text.txt")
data = loader.load()
data
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
splits = text_splitter.split_documents(data)
```

```python
embedding=OllamaEmbeddings()
vectordb=Chroma.from_documents(documents=splits, embedding=embedding)
vectordb
```

```python
query="From where the scent is comming??"
docs=vectordb.similarity_search(query)
docs[0].page_content
```

```python
vectordb=Chroma.from_documents(documents=splits, embedding=embedding, persist_directory="../Data/ignore/chroma_db")
```

```python
db2 = Chroma(persist_directory="../Data/ignore/chroma_db", embedding_function=embedding)
docs= db2.similarity_search(query)
print(docs[0].page_content)
```

```python
docs=vectordb.similarity_search_with_score(query)
docs
```

```python
retriever=vectordb.as_retriever()
retriever.invoke(query)[0].page_content
```
